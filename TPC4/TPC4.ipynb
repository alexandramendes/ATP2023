```python
print("Bem vindo!!!")
print("Menu\nEscolha uma funcionalidade:\n(1) Criar Lista\n(2) Ler Lista\n(3) Soma\n(4) Média\n(5) Maior\n(6) Menor\n(7) estaOrdenada por ordem crescente\n(8) estaOrdenada por ordem decrescente\n(9) Procura um elemento\n(0) Sair")

import random
lista = []
programa = True
while programa:
    n = int(input("Escolha uma funcionalidade:"))
    if n==0:
        programa = False
        print(lista)
        print("Obrigada por jogares!")
    if n == 1:
        i = 0
        while i <= 100:
            lista.append(random.randint(0,100))
            lista.sort()
            i = i+1
        print(lista)
    if n == 2:
        lista.clear()
        tam = int(input("Escolha quantos números quer na sua lista:"))
        while tam > 0:
            num = int(input("Escolha os numeros que quer introduzir"))
            lista.append(num)
            tam = tam - 1
        print(lista)
    if n == 3:
        soma = 0
        for i in lista:
            soma = soma + i
        print(soma)
    if n == 4:
        soma = 0
        for i in lista:
            soma = soma + i
        media = soma/len(lista)
        print(media)
    if n == 5:
        maior = lista[0]
        for i in lista:
            if i > maior:
                maior = i
        print(maior)
    if n == 6:
        menor = lista[0]
        for i in lista:
            if i < menor:
                menor = i
        print(menor)
    if n == 7:
        listacop = lista.copy()
        listacop.sort()
        if lista == listacop:
            print("Sim")
        else:
            print("Não")
    if n == 8:
        listainv = lista.copy()
        listainv.sort()
        listainv.reverse()
        if lista == listainv:
            print("Sim")
        else:
            print("Não")
    if n == 9:
        proc = int(input("Escolha o elemento que quer procurar:"))
        if proc in lista:
            print(f"A posição do elemento escolhido é {lista.index(proc)+1}.")
        else:
            print("-1")

```
