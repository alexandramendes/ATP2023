### TPC6: Análise de dados: doença cardíaca

Descarregue o ficheiro de dados: [diabetes_prediction_dataset.csv](./datasets/diabetes_prediction_dataset.csv.zip).

Encontra mais informação sobre este conjunto de dados [aqui](https://www.kaggle.com/datasets/iammustafatz/diabetes-prediction-dataset).

Crie um programa em Python, conjunto de funções, que responda às seguintes questões:
* Crie uma função que lê a informação do ficheiro para um modelo, previamente pensado em memória;
* Crie uma função que calcula a distribuição da doença por sexo (tenha atenção que indivíduos doentes e não doentes no dataset);
* Crie uma função que calcula a distribuição da doença por escalões etários. Considere os seguintes escalões: [0-10], [11-24], [25-29], [30-34], [35-39], [40-44], ...
* Crie uma função que calcula a distribuição da doença por níveis de glucose. Considere um nível igual a um intervalo de 10 unidades, comece no limite inferior e crie os níveis necessários até abranger o limite superior;
* Crie uma função que imprime na forma de uma tabela uma distribuição;
* Especifique um programa que ao executar apresenta as tabelas correspondentes às distribuições pedidas.

```python
import csv
ficheiro = 'diabetes_prediction_dataset.csv'

def leficheiro(ficheiro):
    info = []
    f = open(str(ficheiro),mode='r')
    for linha in f:
        doente = linha.split(",")
        info.append(doente)
    return info


def dist_doenca(info):
    contadorM = 0
    contadorF = 0
    for pessoa in info:
        if pessoa[0] == 'Female' and pessoa[-1] == '1':
            contadorF = contadorF + 1
        if pessoa[0] == 'Male' and pessoa[-1] == '1':
            contadorM = contadorM + 1
    distrib = [('Masculino',contadorM),('Feminino',contadorF)]
    return distrib


def dist_idade(info):
    distrib = []
    idade = 0
    while idade < 95:
        escalao = [idade, idade + 4]
        contador = 0
        for pessoa in info:
            if pessoa[-1] == '1'  and escalao[0] <= float(pessoa[1]) < escalao[1]:
                contador = contador + 1
        distrib.append((f'{escalao[0]}->{escalao[1]}',contador))
        idade = idade + 5
    return distrib


def dist_glic(info):
    glucose = 0
    distrib = []
    while glucose < 350:
        nivel = [glucose, glucose + 10]
        contador = 0
        for pessoa in info:
            if pessoa[-1] == '1' and nivel[0] <= float(pessoa[-2]) < nivel[1]:
                contador = contador + 1
        distrib.append((f'{nivel[0]}->{nivel[1]}',contador))
        glucose = glucose + 10
    return distrib


def tabela(dados,critério):
    if critério==dist_doenca:
        print(" Sexo  | Nº de pessoas doentes")
    if critério==dist_idade:
        print("Distribuição etária     | Nº de pessoas doentes")
    if critério==dist_glic:
        print("Diabetes | Nº de doentes")
    for i in critério(dados):
        print(i[0], "|", i[1])


def menu():
    print("""Bem-vindo!!! 
Menu:
    (1) Consultar distribuição de doença por sexo
    (2) Consultar distribuição de doença por faixa etária 
    (3) Consultar distribuição de doença por nível de glicose no sangue
    (0) Sair """)
    return

menu()
option=int(input("O que pretende fazer?"))
print("")

dados=leficheiro('diabetes_prediction_dataset.csv')
while option !=0:
    if option == 1:
        tabela(dados[1:],dist_doenca)
        print("")

    if option == 2:
        tabela(dados[1:],dist_idade)
        print("")

    if option == 3:
       tabela(dados[1:],dist_glic)
       print("")

    option = int(input("O que pretende consultar de seguida?"))
print("")
print("Saiu do programa! Volte sempre :)")
print("")
```
