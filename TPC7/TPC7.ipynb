```python
def derivarPolinomio(p):
    for termo in p:
        coeficiente, grau = termo
        if grau == 1:
            print('+',coeficiente,end='')
        if grau != 0:
            print('+',grau*coeficiente,'x ^',grau-1,end='')
        else:
            print('+',0,end='')
    return 

def somaPoli(p1,p2):
    polnovo = []
    for i in p1:
        for j in p2:
            if i[1]==j[1]:
                polnovo.append(p1[i][0] + p2[j][0])
            else:
                polnovo.append(p1[i])
                polnovo.append(p2[j])
    return polnovo
```

```python
tabela = []

def criaPolIn():
    resPol = []
    grau = int(input("Escolha o grau do seu polinomio"))
    while grau >= 0:
        coeficiente = int(input("Escolha o coeficiente que deseja inserir por ordem decrescente de grau:"))
        if coeficiente != 0:
            polinomio = (coeficiente,grau)
            resPol.append(polinomio)
        grau = grau - 1
    return resPol 


def lerPoli(ficheiro):  #voltar a este
    text = []
    f = open(ficheiro, 'r')
    for line in f:
        termos = line.strip().split()
        polinomio = ""
        for term in termos:
                coef_str, expo_str = term.split(";")
                coef = float(coef_str)
                expo = float(expo_str)
                if polinomio:
                    polinomio += " + "
                if expo !=0:
                    polinomio += str(coef) + " x^" + str(expo)
                else:
                    polinomio += str(coef)
        text.append(polinomio)
    f.close()
    return text

def listarPoli(polinomios): 
    print('Número de ordem | Polinómio')
    for i in range(1,len(polinomios)+1):
        tab = (i,polinomios[i-1])
        tabela.append((i,polinomios[i-1]))
        print(f'{tab[0]:^15} {tab[1]}')
    

def calcPoli(p,x):      
    valor = 0
    for termo in p:
        coeficiente,grau = termo
        res = coeficiente*(x**grau)
        valor = valor + res
    return valor

def listarPoliGrau(tabela): 
    numOrdem=1
    print("Numero de ordem |   Polinomio   |        Grau polinomio")
    for i in range(0,len(tabela)):
        print(numOrdem, "   |   ", listarPoli(tabela[i]),"       |   ",grauPolinomio(tabela[i]))
        numOrdem=numOrdem+1
        

def grauPolinomio(p):
    maior = p[0][1]
    for termo in p:
        if termo[1] > maior:
            maior = termo[1]
    return maior

def grauPolinomios(listap):
    res = 0
    for p in listap:
        grau = grauPolinomio(p)
        if grau > res:
            res = grau
    return (p,res)


import matplotlib.pyplot as plt
import numpy as np

def graphPoli(polinomial, intervalo, num_pontos=100):
    x_min, x_max = intervalo
    step = (x_max - x_min) / (num_pontos - 1)
    
    x = [x_min + i * step for i in range(num_pontos)]
    
    y = [0] * num_pontos
    for coef, expo in polinomial:
        for i in range(num_pontos):
            y[i] += coef * (x[i] ** expo)
    
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Polinómio')
    plt.show()


def guardaPoli(t, fnome):
    file = open(fnome, mode = 'w')
    for p in t:
        for termo in p:
            coe, grau=termo
            file.write(str(coe)+";"+ str(grau)+"  ")
        file.write("\n")
    file.close()
    return


```
